### Segmenting and Clustering Neighborhoods in Toronto

```python
import pandas as pd
import numpy as np
import ssl
import urllib.request

print("Hello Capstone Project Course!. lets go")
```

### Use the Notebook to build the code to scrape the following Wikipedia page

Build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe

```python
df=None
context = ssl._create_unverified_context()

with urllib.request.urlopen("http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", context=context) as response:
    html = response.read()
    df = pd.read_html(html, skiprows=[0])[0]
    df.columns = ['Postcode','Borough','Neighbourhood']
    
    #  Ignore cells with a borough that is Not assigned
    df = df[df['Borough']!='Not assigned'].reset_index()
#     display(df.head())

# More than one neighborhood can exist in one postal code area
# These two rows will be combined into one row with the neighborhoods separated with a comma

df=df.groupby('Postcode',as_index=False).agg(lambda x: ', '.join(set(x.dropna())))
df
```

#### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
df['Neighbourhood'].loc[df['Neighbourhood'] == 'Not assigned'] = df['Borough']
df
```

### Use the Geospatial_Coordinates provided to associate lat long to postal code

```python
dfCoord=pd.read_csv("Geospatial_Coordinates.csv")
dfCoord.head()
```

### Join the Geospatial_Coordinates dataframe and the Toronto (Postcode, Borough, Neighbourhood) dataframe on Postal Code

```python
dfNew = pd.merge(dfCoord,df,left_on='Postal Code',right_on='Postcode',how='inner').drop(['Postal Code'], axis = 1) 
dfNew = dfNew[['Postcode', 'Borough', 'Neighbourhood','Latitude', 'Longitude']]
dfNew.rename(columns = {'Postcode':'Postal Code'}, inplace = True)
dfNew
```

### # Filter boroughs containing the word Toronto

```python
dfToronto=dfNew[dfNew['Borough'].str.contains("Toronto")]
dfToronto
```

### Get the geograpical coordinate of Toronto

```python
#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values


address = 'Toronto, Can'

geolocator = Nominatim(user_agent="Toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

### Create map of Toronto using latitude and longitude values

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(dfToronto['Latitude'], dfToronto['Longitude'], dfToronto['Borough'], dfToronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

### Explore and cluster the neighborhoods in Toronto

Used K=4 here

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans

k=4
toronto_clusters = dfToronto.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clusters)
kmeans.labels_
dfToronto.insert(0, 'Cluster Labels', kmeans.labels_)
dfToronto
```

### Generate map to visualize neighborhoods of Toronto and how they cluster together

```python
# create map
import matplotlib.cm as cm
import matplotlib.colors as colors

map_clusters = folium.Map(location=[43.6607002, -79.3850889],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(dfToronto['Latitude'], dfToronto['Longitude'],dfToronto['Neighbourhood'], dfToronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

### As you can see the clusters are quite distinctly separated with K=4 into Downtown, East, West and Central Toronto
